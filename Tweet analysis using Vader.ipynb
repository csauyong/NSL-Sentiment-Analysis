# Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime as dt
import time
```

# Set Up Data Pipelines and Clean Data

```python
tweets = pd.read_csv("HK_df.csv", encoding = "ISO-8859-1")
```

```python
tweets.drop(columns = 'Unnamed: 0')
tweets.drop(columns = 'Country')
```

```python
tweets['Date'] = pd.to_datetime(tweets['Date'])
```

```python
tweets.head(3)
```

# Analysis with Vader

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
sentiment = tweets['Text'].apply(lambda x: analyzer.polarity_scores(x))
tweets = pd.concat([tweets,sentiment.apply(pd.Series)],1)
```

```python
tweets
```

```python
tweets.describe()
```

## Change in sentiment

```python
tweets.sort_values(by='Date', inplace=True)
tweets.index = pd.to_datetime(tweets['Date'])
tweets['mean'] = tweets['compound'].expanding().mean()
tweets['rolling'] = tweets['compound'].rolling('6h').mean()
```

```python
fig = plt.figure(figsize=(20,5))
ax = fig.add_subplot(111)
ax.scatter(tweets['Date'],tweets['compound'], label='Tweet Sentiment')
ax.plot(tweets['Date'],tweets['rolling'], color ='r', label='Rolling Mean')
ax.plot(tweets['Date'],tweets['mean'], color='y', label='Expanding Mean')
ax.set_xlim([dt.date(2020,7,5),dt.date(2020,7,13)])
ax.set(title='Sentiment of Tweets over Time', xlabel='Date', ylabel='Sentiment')
ax.legend(loc='best')
plt.show()
```

## Overall sentiment

```python
tweets['vader_polarity'] = 'Neutral'
tweets.loc[tweets['pos']>tweets['neg']+0.05, 'vader_polarity'] = 'Positive' 
tweets.loc[tweets['neg']>tweets['pos']+0.05, 'vader_polarity'] = 'Negative'
```

```python
ax = tweets['vader_polarity'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Overall sentiment")
ax.set_xlabel("Kind of sentiment")
ax.set_ylabel("Frequency")
```
